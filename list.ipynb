```python
import re 

words = yukcheon['단어']
words_with_numbers = [word for word in words if re.search(r'[0-9]', word)]
print(words_with_numbers)
```

There are several ways to "delete" parts of a string in Python, but it's important to remember a key concept: strings in Python are immutable. This means you can't directly modify an existing string. Instead, when you "delete" parts of a string, you're actually creating a new string with the desired parts removed.
# Regular Expressions:
For more complex pattern-based deletions (like removing all numbers, special characters, or specific patterns), regular expressions are incredibly powerful.

`re.sub(pattern, repl, string)`

```python
yukcheon['단어 Revised'] = ''
```

```python
for word in words:
    word_idx = yukcheon.index[yukcheon['단어'] == word][0]
    #if re.search(r'[0-9]', word):
    nono = re.sub(r'[0-9]', '', word)
    #print(nono)
    yukcheon.loc[word_idx, '단어 Revised'] = nono
```
